```python
dataset4=pd.read_csv("https://github.com/andvise/DataAnalyticsDatasets/blob/f4c1e07915ddfe98f0f5434ec3f0e7f3900f35ab/titanic.csv?raw=true")
dataset4=dataset4.dropna()
print(len(dataset4))
print(dataset4.columns)
print(dataset4)
```

```python
import plotly.express as px
fig = px.bar(dataset4, x='Pclass', y='Survived',color='Pclass')
fig.show()
```

```python
import plotly.graph_objects as go
data_temp1=dataset4.groupby(['Pclass','Sex'])['Name'].count().reset_index()
data_temp1.columns=["Source","Target","Values"]
print(data_temp1)
data_temp1.Source=data_temp1.Source.map({1:"Pclass1",2:"Pclass2",3:"Pclass3"})
data_temp1
```

```python
data_temp2=dataset4.groupby(['Sex','Survived'])['Name'].count().reset_index()
data_temp2.columns=["Source","Target","Values"]
print(data_temp2)
data_temp2.Target=data_temp2.Target.map({1:"Survived",0:"Died"})
data_temp2
```

```python
link=pd.concat([data_temp1,data_temp2],axis=0)
print(link)
unique_source_target=list(pd.unique(link[["Source","Target"]].values.ravel("K")))
mapping_dict={k:v for v,k in enumerate(unique_source_target)}
#mapping_dict
link["Source"]=link["Source"].map(mapping_dict)
link["Target"]=link["Target"].map(mapping_dict)
link
# link_dict=
# link_dict
link_dict=link.to_dict(orient='list')
```

```python
fig=go.Figure(data=[go.Sankey(node=dict(
    pad=10, thickness=20,line=dict(color="blue",width=0.2),label=unique_source_target,color="purple"
),
link=dict(source=link_dict["Source"],
          target=link_dict["Target"],
          value=link_dict["Values"]))])
fig.update_layout(title_text="Titanic Survival Sankey Diagram",font_size=15)
fig.show()
```

```python
import plotly.express as px
fig = px.treemap(dataset4, path=[px.Constant("all"), 'Pclass', 'Sex'], values='Survived',color='Sex')
fig.update_traces(root_color="lightgrey")
fig.update_layout(margin = dict(t=50, l=25, r=25, b=25))
fig.update
fig.show()
```
